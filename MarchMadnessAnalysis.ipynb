```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import warnings
import math
```

```python
warnings.filterwarnings('ignore')
data = pd.read_csv('NCAATourneyCompactResults.csv')
```

```python
win_m = np.mean(data.WScore)
loss_m = np.mean(data.LScore)

print('Average winning score of all games: ' + str(math.floor(win_m)))
print('Average losing score of all games: ' + str(math.floor(loss_m)))

fig = plt.figure()
plt.title('Wining Scores vs Losing Scores')
s = fig.add_subplot(111)
s.bar(['Avg. Losing Score', 'Avg. Winning Score'], [loss_m, win_m], color=['red', 'green'], width=.5)
plt.show()
```

```python
#what percent of games go into overtime?

ot_num = len(data[data.NumOT > 0])

percent_ot = ot_num/len(data.WScore) * 100
print('Percentage of Games Into Overtime: ' + str(percent_ot))

fracs = [ot_num, len(data.WScore) - ot_num]
labels = ['Overtime Games', 'Regular Games']
color = ['orange', 'grey']
explode = (0, 0.05)

plt.pie(fracs, labels=labels, shadow=False, explode=explode, colors=color)
plt.title('Games Into Overtime')
plt.show()
```

```python
sns.distplot(data.WScore, color='green')
sns.distplot(data.LScore, color='red')
plt.title('Losing & Winning Score Distributions')
plt.xlabel('')
plt.show()
```

```python
ot_data = data[data.NumOT > 0]

print("Average winning score in overtime: " + str(math.floor(np.mean(ot_data.WScore))))
print("Average losing score in overtime: " + str(math.floor(np.mean(ot_data.LScore))))

sns.distplot(ot_data.WScore, color='green')
sns.distplot(ot_data.LScore, color='red')
plt.xlabel('')
plt.title('Overtime Losing & Winning Score Distributions')
plt.show()
```
